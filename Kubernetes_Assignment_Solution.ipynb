# Kubernetes Assignment Solution


## Assignment 1
Deploy a 3-node Kubernetes cluster and create an NGINX deployment with 3 replicas.

```python
# Create an NGINX deployment
kubectl create deployment nginx-deployment --image=nginx

# Scale deployment to 3 replicas
kubectl scale deployment nginx-deployment --replicas=3

# Verify
kubectl get deployments
kubectl get pods -o wide
```

## Assignment 2
Expose the deployment using a NodePort service and verify it in a browser.

```python
# Expose deployment
kubectl expose deployment nginx-deployment \
  --type=NodePort \
  --port=80

# View service
kubectl get svc

# Access
minikube service nginx-deployment
```

## Assignment 3
Scale the deployment from 3 replicas to 5 replicas.

```python
kubectl scale deployment nginx-deployment --replicas=5

kubectl get pods
```

## Assignment 4
Change the service type from NodePort to ClusterIP.

```python
kubectl delete service nginx-deployment

kubectl expose deployment nginx-deployment \
  --type=ClusterIP \
  --port=80

kubectl get svc
```

## Assignment 5
Deploy NGINX with 3 replicas, expose it using a ClusterIP service, and create an Ingress.

```python
# Deployment
kubectl create deployment nginx-ingress --image=nginx
kubectl scale deployment nginx-ingress --replicas=3

# Service
kubectl expose deployment nginx-ingress \
  --type=ClusterIP \
  --port=80

# ingress.yaml
apiVersion: networking.k8s.io/v1
kind: Ingress
metadata:
  name: nginx-ingress
spec:
  rules:
  - host: demo.local
    http:
      paths:
      - path: /
        pathType: Prefix
        backend:
          service:
            name: nginx-ingress
            port:
              number: 80
```

```python
kubectl apply -f ingress.yaml

kubectl get ingress

# Add demo.local to /etc/hosts (or Windows hosts file)
# <Minikube-IP> demo.local
```
