## Key features of Python

**Answer:**

Python is simple, readable, interpreted, object-oriented, portable, open-source, dynamically typed, and has a rich standard library.

## Predefined keywords

**Answer:**

Keywords are reserved words with predefined meanings such as `if`, `for`, `while`, `def`, `class`, `return`, and `import`.

## Mutable vs Immutable

**Answer:**

Mutable objects (list, dict, set) can be modified. Immutable objects (int, float, tuple, string) cannot be modified after creation.

## Operators

**Answer:**

Python supports arithmetic, comparison, logical, assignment, bitwise, membership, and identity operators.

## Type Casting

**Answer:**

Type casting converts one data type into another using functions like `int()`, `float()`, `str()`, and `list()`.

## Conditional Statements

**Answer:**

`if`, `elif`, and `else` allow decision-making based on conditions.

## Loops

**Answer:**

Python provides `for` loops for iteration over sequences and `while` loops for repeated execution until a condition becomes false.

### Example: Keywords Example

```python
age=18
if age>=18:
    print('Eligible to vote')
```

### Example: Mutable vs Immutable

```python
lst=[1,2]
lst.append(3)
print(lst)

s='Hello'
s=s+' World'
print(s)
```

### Example: Operators

```python
a,b=10,3
print(a+b)
print(a>b)
print(a and b)
print(a%b)
```

### Example: Type Casting

```python
num='100'
print(int(num)+50)
print(float(num))
```

### Example: Conditional Statement

```python
marks=75
if marks>=90:
    print('A+')
elif marks>=60:
    print('A')
else:
    print('B')
```

### Example: For Loop

```python
for i in range(1,6):
    print(i)
```

### Example: While Loop

```python
count=1
while count<=5:
    print(count)
    count+=1
```
