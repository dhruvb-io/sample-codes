## Difference between function and method

A function is an independent block of code. A method is a function associated with an object or class.

Example:
```python
len([1,2])   # function
'abc'.upper() # method
```

## Function arguments and parameters

Parameters are variables in the function definition. Arguments are the actual values passed during the function call.

### Sum of even numbers

````
def sum_even(nums):
    return sum(n for n in nums if n%2==0)

print(sum_even([1,2,3,4,5,6]))
````

### Reverse string

````
def reverse_string(s):
    return s[::-1]
print(reverse_string("Python"))
````

### Squares of list

````
def squares(nums):
    return [n*n for n in nums]
print(squares([1,2,3,4]))
````

### Prime numbers

````
def is_prime(n):
    if n<2:return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:return False
    return True
print([i for i in range(1,201) if is_prime(i)])
````

### Fibonacci iterator

````
class Fibonacci:
    def __init__(self,n):
        self.n=n;self.a=0;self.b=1;self.i=0
    def __iter__(self): return self
    def __next__(self):
        if self.i>=self.n: raise StopIteration
        self.a,self.b=self.b,self.a+self.b
        self.i+=1
        return self.a
print(list(Fibonacci(10)))
````

### Power of 2 generator

````
def powers(exp):
    for i in range(exp+1):
        yield 2**i
print(list(powers(5)))
````

### Read file generator

````
def read_lines(file):
    with open(file) as f:
        for line in f:
            yield line.strip()
````

### Lambda sort

````
data=[('a',3),('b',1),('c',2)]
print(sorted(data,key=lambda x:x[1]))
````

### Map Celsius to Fahrenheit

````
temps=[0,20,37]
print(list(map(lambda c:c*9/5+32,temps)))
````

### Filter vowels

````
text='Hello World'
print(''.join(filter(lambda c:c.lower() not in 'aeiou',text)))
````
